## Introduction

### There are two main reasons for creating visuals using data:

1. <b>Exploratory analysis "EDA"</b> is done when you are searching for insights. These visualizations don't need to be perfect. You are using plots to find insights, but they don't need to be aesthetically appealing. You are the consumer of these plots, and you need to be able to find the answer to your questions from these plots.

2. <b>Explanatory analysis</b> is done when you are providing your results for others. These visualizations need to provide you the emphasis necessary to convey your message. They should be accurate, insightful, and visually appealing.

**The five steps of the data analysis process:**

1. <b>Extract</b> - Obtain the data from a spreadsheet, SQL, the web, etc.

2. <b>Clean</b> - Here we could use exploratory visuals.

3. <b>Explore</b> - Here we use exploratory visuals.

4. <b>Analyze</b> - Here we might use either exploratory or explanatory visuals.

5. <b>Share</b> - Here is where explanatory visuals live.

## Univariate visualizations
1.  <b>Bar chart</b> for qualitative variables, for example, the amount (number) of eggs consumed in a meal (categories: {breakfast, lunch, or dinner}). In general, bar chart maps categories to numbers.
2. <b>Pie charts</b> is a common univariate plot type that is used to depict relative frequencies for levels of a categorical variable. A pie chart is preferably used when the number of categories is less, and you'd like to see the proportion of each category.
3. <b>Histograms</b> for quantitative variables. A histogram splits the (tabular) data into evenly sized intervals and displays the count of rows in each interval with bars. A histogram is similar to a bar chart, except that the "category" here is a range of values.

```python
# Necessary imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Read the csv file, and check its top 10 rows
df = pd.read_csv('pokemon.csv')
print(df.shape) #(rows,columns)
df.head(10)
```

# 1.1 Bar chart

## Example 1.1.1 Create a vertical bar chart using Matplotlib, with a uniform single color


```python
# Return the Series having unique values
x = df['generation_id'].value_counts(sort=False).index
print(x)

# Return the Series having frequency count of each unique value
y = df['generation_id'].value_counts(sort=False)
print(y)
plt.bar(x, y)

# Labeling the axes
plt.xlabel('generation_id')
plt.ylabel('count')
plt.title('Frequency of generation_id')


# Dsiplay the plot
plt.show()
```

## Example 1.1.2 Create a vertical bar chart using Seaborn, with default colors


```python
sb.countplot(data=df, x='generation_id')
plt.show()
```

## Example 1.1.3 Create a vertical bar chart using Seaborn, with a uniform single color

```python
sb.color_palette()
```

```python
# The `color_palette()` returns the the current / default palette as a list of RGB tuples.
# Each tuple consists of three digits specifying the red, green, and blue channel values to specify a color.
# Choose the first tuple of RGB colors
base_color = sb.color_palette()[0]

# Dynamic-ordering the bars
# The order of the display of the bars can be computed with the following logic.
# Count the frequency of each unique value in the 'generation_id' column, and sort it in descending order
# Returns a Series
freq = df['generation_id'].value_counts()

# Get the indexes of the Series
gen_order = freq.index

# Plot the bar chart in the decreasing order of the frequency of the `generation_id`
sb.countplot(data=df, x='generation_id', color=base_color, order=gen_order)
plt.show()



# Static-ordering the bars
#sb.countplot(data=df, x='generation_id', color=base_color, order=[5,1,3,4,2,7,6]);
#plt.show()
```

# 1.2 Pie chart

## Example 1.2.1 Plot a simple Pie chart

```python
# Use the same pokemon dataset
sorted_counts = df['generation_id'].value_counts() #count the occurrences
print(sorted_counts)
gen_percentage = sorted_counts/sorted_counts[:].sum() #[:] it creates copy of the entire series
plt.pie(gen_percentage, labels = sorted_counts.index, autopct='%1.1f%%')
plt.show()
```

## Example 1.2.2 Plot a simple Donut plot

```python

plt.pie(gen_percentage, labels = sorted_counts.index, wedgeprops = {'width' : 0.6},autopct='%1.1f%%')
plt.show()
```

# 1.3 Histograms

## Example 1.3.1 Histogram with dynamic number of bins

```python
#Histogram with fixed number of bins
plt.hist(data = df, x = 'speed', bins = 10,edgecolor='black', alpha=0.7);
```

```python
# Create bins with step-size 5
#np.arange(start, stop, step)
bins = np.arange(0, df['speed'].max()+5, 5) #bin_size=5
plt.hist(data = df, x = 'speed', bins = bins,edgecolor='black', alpha=0.7);
```

## Example 1.3.2 Plot the similar histogram with Seaborn's histplot()

```python
# use histplot()
sb.histplot(df['speed'], kde=True);
```

# 2.3 Clustered Bar Charts

```python
df2.fuelType.unique()
```

```python
df2.trans.unique()
```

```python
df2['trans_type'] = df2['trans'].apply(lambda x:x.split()[0])
# Applies lambda function to each row in the 'trans' column.
df2.head()
```

```python
sb.countplot(data = df2, x = 'VClass', hue = 'trans_type')
plt.xticks(rotation=45);
```

```python
# Alternative Approach, use heat map using crosstab

sb.heatmap(pd.crosstab(df2.VClass, df2.trans_type), annot=True, fmt='d');
plt.xlabel("Transmission Type")
plt.ylabel("Vehicle Class")
plt.title("Vehicle Class vs Transmission Type Heatmap");
# annot = True makes it so annotations show up in each cell
# Adding fmt = 'd' means that annotations will all be formatted as integers
# You can use fmt = '.0f' if you have any cells with no counts, in order to account for NaNs.
```
